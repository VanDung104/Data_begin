## Bai 2
<font size=5>V·∫Ω ƒë·ªì th·ªã h√†m sin,cos,tan trong kho·∫£ng -2ùúã -> 2ùúã</font>

```python
import matplotlib.pyplot as plt
```

```python
# data
x = np.linspace(-2*np.pi, 2*np.pi, 50)
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)

#draw
plt.plot(x, y_sin, label='sin(x)')
plt.plot(x, y_cos, label='cos(x)')
plt.plot(x, y_tan, label='tan(x)')
plt.title('Cac ham sin, cos, tan')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#set_up_data
y = np.linspace(-5, 5, 10000)
#draw
plt.plot(y**2, y, label='x = y^2')
plt.plot(y, y, label='x=y')

plt.title('Do thi ham x = y^2 va x = y')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()
```

### bai4
<font size=5>T·∫°o 1000 d·ªØ li·ªáu ki·ªÉu int ng·∫´u nhi√™u g·∫Øn v·ªõi 1000 ng√†y t·ª´ 09/02/2010.
D√πng ki·ªÉu pd.Series.T√≠nh cumsum() v√† v·∫Ω ƒë·ªì th·ªã.</font>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#set_data
np.random.seed(0)
data = np.random.randint(low=0, high=1000, size = 1000)

dates = pd.date_range(start = '09/02/2010', periods=1000)

series = pd.Series(data, index=dates)

#Cumsum()
cumulative_sum = series.cumsum()

#draw
plt.plot(cumulative_sum)
plt.title('Cumsum()')
plt.xlabel('Date')
plt.ylabel('Cum_sum')

plt.show()
```
