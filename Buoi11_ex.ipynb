```python
# data
x = np.linspace(-2*np.pi, 2*np.pi, 50)
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)

#draw
plt.plot(x, y_sin, label='sin(x)')
plt.plot(x, y_cos, label='cos(x)')
plt.plot(x, y_tan, label='tan(x)')
plt.title('Cac ham sin, cos, tan')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#set_data
np.random.seed(0)
data = np.random.randint(low=0, high=1000, size = 1000)

dates = pd.date_range(start = '09/02/2010', periods=1000)

series = pd.Series(data, index=dates)

#Cumsum()
cumulative_sum = series.cumsum()

#draw
plt.plot(cumulative_sum)
plt.title('Cumsum()')
plt.xlabel('Date')
plt.ylabel('Cum_sum')

plt.show()
```

```python
#!/usr/bin/env python3

import csv
import os
import time
from datetime import datetime


def find_max_price(datafile):
    f = open(datafile)
    dr = csv.DictReader(f, ["time", "price", "UNKNOWN"])  # NOQA
    # Vit tip code vo y

    try:
        # Xo dng sau v vit code vo y set cc gi tr ph hp
        result = max(dr, key=lambda x : float(x['price']))
        date_result = time.gmtime(int(result['time']))
        date_result = str(date_result.tm_year) + \
            '-' + str(date_result.tm_mon) + \
            '-' + str(date_result.tm_mday)
        formatted_date = datetime.strptime(date_result, "%Y-%m-%d"). \
            strftime("%Y-%m-%d")
        formatted_result = (formatted_date, float(result['price']))
    finally:
        f.close()

    return formatted_result


def solve():
    """Tm ngy gi BTC ln cao nht. Tr v Tuple cha ngy  nh dng
    YYYY-mm-dd (VD: 2017-06-19) v gi VND ca 1 BTC
    """
    # http://api.bitcoincharts.com/v1/csv/
    datafile = "localbtcVND.csv"
    exdir = os.path.dirname(__file__)
    datapath = os.path.join(exdir, datafile)

    result = find_max_price(datapath)
    return result


def main():
    now = time.gmtime(int(time.time()))
    print(now.tm_year, now.tm_mon, now.tm_mday)
    print(solve())


if __name__ == "__main__":
    main()
```
